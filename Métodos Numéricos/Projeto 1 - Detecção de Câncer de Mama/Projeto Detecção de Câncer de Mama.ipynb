```python
import numpy as np
from scipy import ndimage
import matplotlib.pyplot as plt
```

```python
#com anomalias
image_path = 'C:/Users/jacks/OneDrive/Área de Trabalho/metodos/10272_idx5_x1651_y1001_class1.png'  # Substitua pelo caminho da imagem real
image = plt.imread(image_path)
```

```python
gray_image = np.mean(image, axis=2)
smoothed_image = ndimage.gaussian_filter(gray_image, sigma=2)
threshold = 0.5
binary_mask = smoothed_image > threshold
highlighted_image = np.copy(image)
highlighted_image[~binary_mask] = [1, 0, 0]  # Realçar os pixels não mascarados (vermelho)
plt.figure(figsize=(10, 5))
plt.subplot(1, 2, 1)
plt.imshow(image)
plt.title('Imagem Original Paciente 1')
plt.subplot(1, 2, 2)
plt.imshow(highlighted_image)
plt.title('Artefatos Anormais Realçados Paciente 1')
plt.show()
```
