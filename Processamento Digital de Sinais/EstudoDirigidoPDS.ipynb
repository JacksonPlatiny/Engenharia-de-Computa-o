Propriedades da ROC

No processamento digital de sinais, assume-se que os sinais são causais, uma vez que quase todos os dados digitais são adquiridos em tempo real. Portanto, a única ROC de interesse para nós é aquela mencionada na declaração 2.

2. A sequência x1(n) = a^n u(n) é um caso especial de uma sequência à direita, definida como uma sequência x(n) que é zero para alguns n < n0. A ROC para sequências à direita está sempre fora de um círculo de raio Rx-. Se n0 >= 0, então a sequência à direita também é chamada de sequência causal.
3. A sequência x2(n) = -b^n u(-n-1) é um caso especial de uma sequência à esquerda, definida como uma sequência x(n) que é zero para alguns n > n0. Se n0 <= 0, a sequência resultante é chamada de sequência anti-causal. A ROC para sequências à esquerda está sempre dentro de um círculo de raio Rx+.
4. A sequência x3(n) = x1(n) + x2(n)  é uma sequência de dois lados. A ROC para sequências de dois lados é sempre um anel aberto Rx- < |z| < Rx+, se existir.
5. As sequências que são zero para n < n1 e n > n2 são chamadas de sequências de duração finita. A ROC para tais sequências é todo o plano z. Se n1 < 0, então z = ∞ não está na ROC. Se n2 > 0, então z = 0 não está na ROC.
6. A ROC não pode incluir um polo, uma vez que X(z) converge uniformemente ali.
7. Existe pelo menos um polo na fronteira de uma ROC de X(z) racional.
8. A ROC é uma região contígua; ou seja, a ROC não é fragmentada.

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy import signal

# Parâmetro a
a = 0.5  # Alterar o valor de a conforme necessário (0 < |a| < ∞)

# Criar a sequência x1(n) = a^n u(n)
n = np.arange(0, 21)
x1 = a**n * (n >= 0)

# Calcular polos e zeros
sys = signal.TransferFunction([1], [1, -a])
zeros, poles, _ = signal.tf2zpk(sys.num, sys.den)

# Plotar a sequência
plt.stem(n, x1, basefmt='b', use_line_collection=True)
plt.title('Sequência $x_1(n) = a^n u(n)$ para a = {}'.format(a))
plt.xlabel('n')
plt.ylabel('$x_1(n)$')
plt.show()

# Plotar o diagrama de polos e zeros
plt.scatter(np.real(zeros), np.imag(zeros), marker='o', label='Zeros', color='b')
plt.scatter(np.real(poles), np.imag(poles), marker='x', label='Polos', color='r')
plt.axhline(0, color='black',linewidth=0.5)
plt.axvline(0, color='black',linewidth=0.5)
plt.title('Diagrama de Polos e Zeros')
plt.xlabel('Parte Real')
plt.ylabel('Parte Imaginária')
plt.legend()
plt.grid(True)
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy import signal

# Parâmetros a e b
a = 0.5  # Alterar o valor de a conforme necessário (0 < |a| < ∞)
b = 0.8  # Alterar o valor de b conforme necessário (0 < |b| < ∞)

# Calcular polos e zeros para X1(z)
sys1 = signal.TransferFunction([1], [1, -a])
zeros1, poles1, _ = signal.tf2zpk(sys1.num, sys1.den)

# Calcular polos e zeros para X2(z)
sys2 = signal.TransferFunction([-1], [1, -b])
zeros2, poles2, _ = signal.tf2zpk(sys2.num, sys2.den)

# Calcular polos e zeros para X3(z)
sys3 = signal.TransferFunction([b, (a - 1)], [1, -a])
zeros3, poles3, _ = signal.tf2zpk(sys3.num, sys3.den)

# Plotar o diagrama de polos e zeros
plt.scatter(np.real(zeros3), np.imag(zeros3), marker='o', label='Zeros (X3)', color='b')
plt.scatter(np.real(poles3), np.imag(poles3), marker='x', label='Polos (X3)', color='r')
plt.scatter(np.real(zeros1), np.imag(zeros1), marker='s', label='Zeros (X1)', color='g')
plt.scatter(np.real(poles1), np.imag(poles1), marker='^', label='Polos (X1)', color='purple')
plt.scatter(np.real(zeros2), np.imag(zeros2), marker='d', label='Zeros (X2)', color='c')
plt.scatter(np.real(poles2), np.imag(poles2), marker='v', label='Polos (X2)', color='orange')

plt.axhline(0, color='black', linewidth=0.5)
plt.axvline(0, color='black', linewidth=0.5)
plt.title('Diagrama de Polos e Zeros')
plt.xlabel('Parte Real')
plt.ylabel('Parte Imaginária')
plt.legend()

# Calcular e plotar a ROC
roc_radius1 = 1.1  # Raio ROC para X1
roc_x1 = roc_radius1 * np.cos(np.linspace(0, 2*np.pi, 100))
roc_y1 = roc_radius1 * np.sin(np.linspace(0, 2*np.pi, 100))
plt.plot(roc_x1, roc_y1, linestyle='dashed', color='g', label='ROC1 (X1)')

roc_radius2 = 1.1  # Raio ROC para X2
roc_x2 = roc_radius2 * np.cos(np.linspace(0, 2*np.pi, 100))
roc_y2 = roc_radius2 * np.sin(np.linspace(0, 2*np.pi, 100))
plt.plot(roc_x2, roc_y2, linestyle='dashed', color='c', label='ROC2 (X2)')

roc_radius3 = min(roc_radius1, roc_radius2)  # Interseção das ROCs
roc_x3 = roc_radius3 * np.cos(np.linspace(0, 2*np.pi, 100))
roc_y3 = roc_radius3 * np.sin(np.linspace(0, 2*np.pi, 100))
plt.plot(roc_x3, roc_y3, linestyle='dashed', color='m', label='ROC3 (X3)')

plt.grid(True)
plt.show()
```

Transformada Rápida de Fourier

Embora a DFT seja o melhor procedimento matemático para determinar o conteúdo espectral de uma seqüência no domínio do tempo, ela é muito ineficiente. Em 1965, um artigo foi publicado por J.W.Cooley e J.W.Tukey descrevendo um algoritmo eficiente para implementação da DFT, este algoritmo ficou conhecido como Transformada rápida de Fourier (FFT). Antes do advento da FFT, a DFT com muitos pontos estava restrita a grandes centros de pesquisas. Graças a Cooley e Tukey, e a indústria dos semicondutores, DFTs com 1024 pontos podem ser calculadas em apenas alguns segundos em computadores pessoais.

O esforço computacional pode ser definido como o número máximo de operações elementares necessárias para resolver o problema. No caso da DFT, pode-se tratar da complexidade multiplicativa e complexidade aditiva i.e., número de multiplicações ponto flutuante (respectivamente adições) necessárias para calculá-la. Tradicionalmente tem-se usado apenas a complexidade multiplicativa como o parâmetro mais importante. A FFT foi implementada com o objetivo de diminuir complexidade (temporal) necessária para calcular uma DFT (Transformada Discreta de Fourier), visando aplicações em tempo real.Para uma seqüência de N pontos, o algoritmo comum para cálculo da DFT realiza N2 multiplicações, enquanto o algoritmo FFT realiza apenas (N/2)log2N.

A FFT usa um número reduzido de operações aritméticas para calcular a DFT em relação ao seu cálculo direto. As primeiras aplicações práticas da FFT usando computação digital foram resultantes de manipulações da DFT. Existem diversas variações no algoritmo da FFT. O algoritmo de Cooley-Tukey ou de decimação no tempo re-arranja os elementos de entrada e depois constrói a saída em log2N iterações. Também é possível derivar algoritmos de FFT que primeiro desenvolvam um conjunto de log2N iterações nos dados de entrada e re-arranje os valores de saída em ordem reversa de bits. Tais algoritmos são chamados de decimação em freqüência ou algoritmos de FFT de Sande-Tukey. Para algumas aplicações, como a convolução, os dados são convertidos para o domínio da freqüência e depois de volta ao domínio do tempo. Nesses casos, é possível evitar todo o reverso dos bits.

Outra classe de FFTs sub-dividem o conjunto de dados inicial de comprimento N em conjunto menores de tamanho proporcional a potências de 2, por exemplo, N = 4, FFTs de base 4, ou N = 8, FFTs de base 8. Essas transformações são feitas por seções de códigos otimizadas que usam características de simetria desses valores de N. Por exemplo, para N = 4, os senos e cossenos são todos ±1 ou 0, eliminando muitas multiplicações, deixando mais adições e subtrações. A melhoria de desempenho em relação a FFTs mais simples é de cerca de 20 a 30%.

Existem também algoritmos de FFT que atuam em conjuntos de comprimento N que não é potência de 2. Eles trabalham usando relações análogas ao lema de Danielson-Lanczos para sub-dividir o problema inicial em problemas sucessivamente menores, não por potências de 2, mas por qualquer primo pequeno que divida N. Quanto maior for o maior fator primo de N, pior o método funciona. Se N é primo, então nenhuma sub-divisão é possível e o usuário termina usando uma FFT com complexidade O(N2) ao invés de O(N.log2N). Deve-se ficar atento a esse tipo de implementação.

Algoritmos de Winograd são de certa forma análogos às FFTs de base 4 e 8. Winograd derivou códigos altamente otimizados para atuar em DFTs de N pequeno, e.g., para N = 2, 3, 4, 5, 7, 8, 11, 13, 16. Os algoritmos também usam um método mais eficiente para combinar os sub-fatores. O método envolve uma re-ordenação de dados antes e depois do processamento, conseguindo uma grande redução no número de multiplicações do algoritmo. Para valores de N da ordem de potências de 2, o algoritmo de Winograd pode ser mais rápido que algoritmos mais simples de FFT. Essa vantagem deve ser analisada com cuidado já que o método torna bastante mais complicada a indexação dos dados envolvidos na transformada.

```python
import numpy as np
import matplotlib.pyplot as plt

# Dada sequência x1(n)
x1_n = np.array([6, 5, 4, 3, 2, 1])

# Função para calcular DTFT de x1(n)
def calculate_DTFT(xn, omega_values):
    X_ejw = np.zeros(len(omega_values), dtype=complex)
    for i, omega in enumerate(omega_values):
        X_ejw[i] = np.sum(xn * np.exp(-1j * omega * np.arange(len(xn))))
    return X_ejw

# Função para amostrar DTFT nos pontos ωk
def sample_DTFT(X_ejw, k_values, omega_1):
    return X_ejw[k_values]

# Função para calcular IDFS
def calculate_IDFS(X_tilde_k, N):
    n_values = np.arange(N)
    k_values = np.arange(N)
    WNnk = np.exp(1j * 2 * np.pi / N) ** (-np.outer(n_values, k_values))
    xn_tilde = np.dot(X_tilde_k, WNnk) / N
    return xn_tilde.real

# Parâmetros
N = len(x1_n)
omega_values = np.linspace(0, 2 * np.pi, 1000)
k_values = np.arange(-N // 2, N // 2)

# Calcular DTFT de x1(n)
X1_ejw = calculate_DTFT(x1_n, omega_values)

# Amostrar DTFT nos pontos ωk
omega_1 = 2 * np.pi / 4
X1_sampled = sample_DTFT(X1_ejw, k_values, omega_1)

# Calcular IDFS para obter x̃2(n)
x2_tilde_n = calculate_IDFS(X1_sampled, N)

# Plotar magnitude de X1(e^{jω})
plt.figure(figsize=(12, 6))
plt.subplot(2, 1, 1)
plt.plot(omega_values, np.abs(X1_ejw))
plt.title('Magnitude of X1(e^{jω})')
plt.xlabel('ω')
plt.ylabel('|X1(e^{jω})|')
plt.grid(True)

# Plotar pontos amostrados
plt.subplot(2, 1, 2)
plt.stem(omega_1 * k_values, np.abs(X1_sampled), use_line_collection=True)
plt.title('Amplitude Samples at ωk')
plt.xlabel('ωk')
plt.ylabel('|X1(e^{jωk})|')
plt.grid(True)

plt.tight_layout()
plt.show()

# Exibir resultado da IDFS
print("Inverse DFS Sequence x̃2(n):", x2_tilde_n)
```

1. Somador: Este elemento tem duas entradas e uma saída. Observe que a adição de três ou mais sinais é implementada por somadores sucessivos de duas entradas.

```python
import numpy as np

def par2dir(C0, B, A):
    num_secoes = len(B)

    # Inicializar numerador e denominador geral
    b_direto = B[0]
    a_direto = A[0]

    # Iterar sobre as seções paralelas e convoluir
    for i in range(1, num_secoes):
        b_direto = np.convolve(b_direto, A[i]) + np.convolve(a_direto, B[i])
        a_direto = np.convolve(a_direto, A[i])

    # Adicionar termo comum
    b_direto = C0 * np.convolve(b_direto, [1, 0])

    return b_direto, a_direto

# Definir parâmetros para as seções paralelas
C0 = 0

B1 = np.array([2, 4])
A1 = np.array([1, 1, 0.9])

B2 = np.array([0.5, 0.7])
A2 = np.array([1, -1, 0.8])

# Converter seções paralelas para forma direta
b_direto, a_direto = par2dir(C0, [B1, B2], [A1, A2])

# Exibir coeficientes da forma direta geral
print("Coeficientes da Forma Direta Geral:")
print("Numerador:", b_direto)
print("Denominador:", a_direto)
```

Introdução

Em processamento digital de sinais há dois importantes tipos de
sistemas: o primeiro tipo filtra os sinais no domínio do tempo e são chamados
filtros digitais. O segundo tipo provê a representação do sinal no domínio da
frequência e são chamados de analisadores de espectro.
Os filtros são projetados como seletores de frequência. Há considerações
diferentes se o projeto é voltado para criação de filtros FIR ou IIR.

O projeto de um filtro digital tem três passos:

• Especificações: determinada pela aplicação a qual o filtro está sendo
construído;
• Aproximações: o projeto do filtro especificamente. Aqui, define-se H(z).
• Implementações: a transcrição da função de transferência para hardware
ou software.

Em diversas aplicações como processamento de voz ou som, filtros digitais são
usados para implementar operações seletivas de frequência. Assim,
especificações são necessárias no domínio da frequência em termos de
magnitude desejada e resposta em fase do filtro. Em geral, uma resposta em fase linear na banda de passagem é necessária. No caso de filtros FIR, é possível encontrar um filtro de
fase exatamente linear. Para filtros IIR, uma fase linear na banda de passagem
não é possível. Assim, consideraremos especificações em magnitude apenas.

Especificações absolutas que provêem um conjunto de requisitos na magnitude

Técnica de Projeto Ótimo Equirriple

Os métodos de janelamento e de amostragem na frequência têm alguns problemas:

1) Não podemos especificar wP e wS precisamente nos projetos.
2) Não podemos especificar δ1 e δ2 simultaneamente. Ou consideramos δ1 = δ2 (como no janelamento) ou otimizamos δ2 (como na amostragem).
3) O erro de aproximação não é distribuído uniformemente nas bandas. Ele é mais alto perto das fronteiras das bandas e menor quanto mais distante delas.

O método equirriple ótimo evita esses problemas. No entanto ele é bastante difícil de utilizar e requer computador na sua implementação. O objetivo é minimizar o erro máximo de aproximação (minimax do erro). Tais filtros são chamados de equirriple porque o erro é distribuído de maneira uniforme na banda de passagem e de corte o que resulta em um filtro de menor ordem.

Introdução

Filtros IIR têm resposta infinita ao impulso, assim eles podem ser igualados a
filtros analógicos, os quais, em geral, têm resposta ao impulso infinitamente

fato que tanto tabelas de filtros analógicos quanto as conversões estão

```python
import numpy as np
import matplotlib.pyplot as plt

# Coeficientes do sistema contínuo
c = [1, 1]
d = [1, 5, 6]

# Tempo de amostragem
T = 0.1

# Função para realizar a invariância de impulso
def imp_invar(c, d, T):
    # Coeficientes do sistema discreto
    b = [c[0] * T]
    a = [1, -np.exp(-c[1] * T)]

    for i in range(2, len(d)):
        b.append(d[i] * T**i)

    for i in range(2, len(d)):
        a.append(-np.exp(-c[1] * T) * T**(i-1))

    return b, a

# Aplicar a função imp_invar para obter os coeficientes do filtro discreto
b, a = imp_invar(c, d, T)

print(f"b = {b}")
print(f"a = {a}")

# Criar um filtro discreto usando os coeficientes
sistema_discreto = (b, a, T)

# Criar um sinal de impulso discreto
n = np.arange(0, 20)
u = np.zeros_like(n)
u[0] = 1

# Aplicar o filtro discreto manualmente
resposta_discreta = np.zeros_like(u)

for i in range(len(n)):
    termo_b = b[0] * u[i]
    termo_a = 0 if i == 0 else a[1] * resposta_discreta[i - 1]

    resposta_discreta[i] = termo_b + termo_a

# Plotar a resposta ao impulso
plt.stem(n, resposta_discreta, use_line_collection=True)
plt.title('Resposta ao Impulso do Filtro Discreto')
plt.xlabel('n')
plt.ylabel('Amplitude')
plt.grid(True)
plt.show()
```

1. Pulse-Code Modulation (PCM):
É amplamente utilizado na transmissão de áudio e vídeo. Em sistemas telefônicos, a voz é frequentemente convertida em PCM para transmissão eficiente. Também é usado em sistemas de gravação de áudio digital.

2. Differential PCM (DPCM):
Redução da redundância na transmissão de sinais de áudio e vídeo. Pode ser aplicado em sistemas de comunicação digital onde a precisão é essencial, como em transmissões de imagens médicas.

3. Adaptive PCM and DPCM (ADPCM):
Usado em compressão de áudio, especialmente em transmissões de voz sobre redes de comunicação. Também encontrado em sistemas de armazenamento de áudio digital.

4. Delta Modulation (DM):
Comum em transmissão de voz de baixa qualidade, como em sistemas de rádio de onda curta. Pode ser utilizado em sistemas de comunicação com largura de banda limitada.

5. Linear Predictive Coding (LPC) of Speech:
Amplamente utilizado na compressão de voz em telefonia móvel e VoIP. Também é aplicado em sistemas de reconhecimento de voz.

6. Dual-Tone Multifrequency (DTMF) Signals:
Usado em telefones para enviar sinais de controle, como na discagem por tons. Também encontrado em sistemas de automação e resposta interativa por voz (IVR).

7. Binary Digital Communications:
Fundamental em todas as formas de comunicação digital, incluindo redes de computadores, transmissões de dados e internet. Pode ser aplicado em sistemas de modulação e demodulação para transferência eficiente de dados.

8. Spread-Spectrum Communications:
Encontrado em sistemas de comunicação sem fio, como redes Wi-Fi e Bluetooth. Também utilizado em comunicações militares e seguras devido à sua resistência a interferências e segurança.

Implementação PCM:

```python
import numpy as np
import matplotlib.pyplot as plt

def encode_pcm(signal, quantization_bits=8):
    # Normalizar o sinal para valores entre -1 e 1
    normalized_signal = np.interp(signal, (signal.min(), signal.max()), (-1, 1))

    # Quantização do sinal
    quantized_signal = np.round((normalized_signal + 1) * (2 ** (quantization_bits - 1)))

    return quantized_signal.astype(int)

def decode_pcm(quantized_signal, quantization_bits=8):
    # Desfazer a quantização
    normalized_signal = (quantized_signal / (2 ** (quantization_bits - 1))) - 1

    return normalized_signal

# Criar um sinal de exemplo (uma onda senoidal)
fs = 1000  # Frequência de amostragem
t = np.arange(0, 1, 1/fs)  # Vetor de tempo
f = 5  # Frequência da onda senoidal
signal = np.sin(2 * np.pi * f * t)

# Codificar o sinal em PCM
quantized_signal = encode_pcm(signal)

# Decodificar o sinal PCM
decoded_signal = decode_pcm(quantized_signal)

# Plotar os resultados
plt.figure(figsize=(10, 6))

plt.subplot(3, 1, 1)
plt.plot(t, signal)
plt.title('Sinal Original')

plt.subplot(3, 1, 2)
plt.stem(t, quantized_signal, basefmt='b')
plt.title('Sinal PCM Quantizado')

plt.subplot(3, 1, 3)
plt.plot(t, decoded_signal)
plt.title('Sinal Decodificado')

plt.tight_layout()
plt.show()
```

Implementação DM:

```python
import numpy as np
import matplotlib.pyplot as plt

def encode_delta_modulation(signal, step_size=0.1):
    encoded_signal = np.zeros_like(signal, dtype=int)

    for i in range(1, len(signal)):
        delta = signal[i] - signal[i-1]
        if delta > step_size:
            encoded_signal[i] = 1
        elif delta < -step_size:
            encoded_signal[i] = -1

    return encoded_signal

def decode_delta_modulation(encoded_signal, step_size=0.1):
    decoded_signal = np.zeros_like(encoded_signal, dtype=float)

    for i in range(1, len(encoded_signal)):
        decoded_signal[i] = decoded_signal[i-1] + encoded_signal[i] * step_size

    return decoded_signal

# Criar um sinal de exemplo (uma onda senoidal)
fs = 1000  # Frequência de amostragem
t = np.arange(0, 1, 1/fs)  # Vetor de tempo
f = 5  # Frequência da onda senoidal
signal = np.sin(2 * np.pi * f * t)

# Codificar o sinal usando Delta Modulation
encoded_signal = encode_delta_modulation(signal)

# Decodificar o sinal Delta Modulation
decoded_signal = decode_delta_modulation(encoded_signal)

# Plotar os resultados
plt.figure(figsize=(10, 6))

plt.subplot(3, 1, 1)
plt.plot(t, signal)
plt.title('Sinal Original')

plt.subplot(3, 1, 2)
plt.step(t, encoded_signal, where='post')
plt.title('Sinal Delta Modulation Codificado')

plt.subplot(3, 1, 3)
plt.plot(t, decoded_signal)
plt.title('Sinal Decodificado')

plt.tight_layout()
plt.show()
```
