##DEFINIÇÃO DO PROJETO:#


O SSD (Self-Service Drink) é um dispositivo que cria drinks a partir de misturas de bebidas. A mistura das bebidas é feita em um recipiente interno, o qual é preenchido com as bebidas-base, que são injetadas no recipiente através de sucção direto da garrafa. A escolha do drink e algumas misturas são escolhas do usuário, através de uma tela interativa.

##1. Descrição Funcional##

####Controle por Aplicativo (MIT App Inventor):

Objetivo: Permitir que os usuários controlem a máquina de drinks remotamente através de um aplicativo intuitivo.

Funcionalidades Principais:
* Seleção de drinks disponíveis.
* Escolha do tamanho do copo.


####Comunicação Bluetooth (Arduino):

Objetivo: Estabelecer uma comunicação sem fio entre o aplicativo e o hardware (Arduino).

Funcionalidades Principais:
* Transmissão de comandos de preparo de drinks do aplicativo para o Arduino.
* Recebimento de feedback do Arduino para atualizações em tempo real no aplicativo.


####Bomba de Ingredientes e Relé:

Objetivo: Dispensar ingredientes precisamente conforme instruções do aplicativo.

Funcionalidades Principais:
* Controle do acionamento das bombas por meio de relés.
* Personalização das quantidades de ingredientes para diferentes receitas.


#### Sensor Ultrassônico:

Objetivo: Evitar derramamentos durante o processo de preparo dos drinks.

Funcionalidades Principais:
* Medição da distância entre o sensor e a superfície do líquido no copo.
* Envio de informações em tempo real para evitar o excesso de despejo.


#### Fonte de Alimentação de 12V:

Objetivo: Fornecer energia elétrica estável para todos os componentes do sistema.

Funcionalidades Principais:
* Conversão de energia da tomada para 12V.
* Alimentação das bombas, Arduino e outros dispositivos eletrônicos.



#### Segurança e Feedback do Usuário:

Objetivo: Garantir a segurança durante a operação e fornecer feedback claro ao usuário.

Funcionalidades Principais:
* Implementação de mecanismos de segurança para evitar operações incorretas.
* Feedback visual e/ou sonoro para indicar o status do processo de preparo.



##2. Descrição Arquitetural##

###Camada de Interface do Usuário (UI):###

Componentes:
* Aplicativo desenvolvido no MIT App Inventor.

Funções Principais:
* Interface gráfica para seleção de drinks, personalização de receitas e escolha do tamanho do copo.
* Comunicação com a camada de Controle para enviar comandos de preparo e receber feedback.

###Camada de Controle:###

Componentes:
* Módulo de Controle central baseado em Arduino.

Funções Principais:
* Recebe comandos do aplicativo por meio da comunicação Bluetooth.
* Controla o acionamento das bombas de ingredientes e relés para garantir a precisão no preparo.
* Recebe dados do sensor ultrassônico para prevenção de derramamento.

###Camada de Sensores e Atuadores:###

Componentes:
* Sensor ultrassônico para prevenção de derramamento.
* Bombas de ingredientes.

Funções Principais:
* Mede a distância para evitar derramamentos.
* Controla o fluxo de ingredientes com precisão.

###Camada de Alimentação:###

Componentes:
* Fonte de alimentação de 12V.

Funções Principais:
* Fornece energia estável para todos os componentes do sistema.

* Usuário Interage com o Aplicativo: Seleciona um drink, personaliza a receita e escolhe o tamanho do copo.

* Aplicativo Envia Comandos para o Arduino: Utiliza a comunicação Bluetooth para enviar os comandos de preparo para o módulo de controle.

* Módulo de Controle Aciona Sensores e Atuadores: Ativa as bombas de ingredientes e controla os relés para garantir a precisão na preparação e monitora o sensor ultrassônico para evitar derramamentos.

##3. Descrição das Tecnologias Utilizadas##

#### MIT App Inventor:

Propósito: O MIT App Inventor é uma plataforma de desenvolvimento visual que permite criar aplicativos Android de forma intuitiva, sem a necessidade de programação extensiva.

Funcionalidades Principais:
* Interface gráfica para design de aplicativos.
* Programação visual por blocos para lógica e controle de fluxo.
* Integração com sensores, câmera, GPS e outros recursos do dispositivo Android.
* Conexão fácil com hardware externo por meio de Bluetooth ou Wi-Fi.



#### Arduino:

Propósito: O Arduino é uma plataforma de hardware open-source que facilita a prototipagem de projetos eletrônicos, sendo amplamente utilizada para automação e controle de dispositivos.

Funcionalidades Principais:
* Placa de microcontrolador para controle de entradas e saídas.
* Diversos pinos de entrada/saída digital e analógica.
* Ambiente de desenvolvimento (IDE) para programação em C/C++.
* Facilidade de interface com sensores, atuadores e outros componentes eletrônicos.



#### Comunicação Bluetooth:

Propósito: Utilizada para estabelecer a comunicação sem fio entre o aplicativo (MIT App Inventor) e o hardware (Arduino), permitindo o controle remoto.

Funcionalidades Principais:
* Transmissão de dados serial entre dispositivos.
* Emparelhamento seguro para garantir a integridade da comunicação.
* Ativação/desativação remota de recursos no hardware.



####Relé para Acionamento de Bombas:

Propósito: O relé é um componente crucial para controlar o acionamento das bombas que dispensam os líquidos da máquina de drinks.

Funcionalidades Principais:
* Atua como um interruptor controlado eletronicamente que permite ou interrompe o fluxo de corrente para as bombas.
* Controlado pelo Arduino, o relé é acionado para iniciar ou interromper o fornecimento de ingredientes durante o processo de preparo do drink.
* Pode ser parte integrante de um circuito de potência para assegurar um controle seguro das bombas de bebida.



#### Sensor Ultrassônico:

Propósito: O sensor ultrassônico desempenha um papel importante na prevenção de derramamento, ajudando a detectar a presença e altura dos copos na área de preparação.

Funcionalidades Principais:
* Utiliza ondas ultrassônicas para medir a distância entre o sensor e a superfície do líquido no copo.
* Permite a identificação precisa da altura do líquido no copo, evitando derramamentos durante o processo de despejo dos ingredientes.
* Integrado ao Arduino, proporciona informações em tempo real sobre a posição dos copos.



####Bomba de Líquidos Controlada por Relé:

Propósito: A bomba é responsável por dispensar ingredientes específicos da máquina de drinks durante o processo de preparo.

Funcionalidades Principais:
* Acionada pelo relé, a bomba é controlada eletronicamente para fornecer a quantidade correta de cada ingrediente.
* Pode ser ajustada para diferentes taxas de fluxo, permitindo a personalização das receitas.
* Integrada ao sistema para fornecer ingredientes com precisão durante a preparação dos drinks.



####Fonte de Alimentação de 12V:

Propósito: A fonte de alimentação de 12V fornece energia elétrica para as bombas de ingredientes, o Arduino e outros componentes eletrônicos necessários para o funcionamento da máquina de drinks.

Funcionalidades Principais:
* Converte a energia elétrica da tomada para uma voltagem adequada (12V) para alimentar as bombas e outros componentes eletrônicos.
* Fornece a potência necessária para garantir o funcionamento eficiente de todos os dispositivos conectados.
* Integrada ao sistema para garantir uma operação estável e confiável da máquina de drinks.


##4. Requisitos Funcionais e Não Funcionais##

##6. Documentação de Código

O código abaixo está presente no Arduino, onde há o controle das bombas e o recebimento dos comandos enviados via bluetooth.

```python
//#include <SoftwareSerial.h>
//SoftwareSerial  EEBlue(0, 1); // RX | TX

// Definindo os pinos utilizados para cada bebida.
int bebida_azul = 2;
int bebida_verde = 3;
int bebida_amarela = 4;
int bebida_vermelha = 5;

// Cada bomba suga 100ml a cada 3 segundos
// 166ml = 5000 milissegundos
// 250ml = 7500 milissegundos
// 500ml = 15000 milissegundos

// Tempo para quando são 3 bombas para sugar os 500ml
int _166MlBebidaAzul = 5000;
int _166MlBebidaVerde = 5000;
int _166MlBebidaAmarela = 5000;
int _166MlBebidaVermelha = 5000;

// Valor em milissegundos do tempo de sucção de cada bomba
int _250MlBebidaAzul = 7500;
int _250MlBebidaVerde = 7500;
int _250MlBebidaAmarela = 7500;
int _250MlBebidaVermelha = 7500;

// Tempo que bombas ficaram ativadas para limpeza
int tempoLimpeza = 8000;

// Quanto tempo existe entre a mudança para a próxima bomba.
int  tempoEntreSuccoes = 500;

int _500MlBebidaAzul = _250MlBebidaAzul*2;
int _500MlBebidaVerde = _250MlBebidaVerde*2;
int _500MlBebidaAmarela = _500MlBebidaAmarela*2;
int _500MlBebidaVermelha = _500MlBebidaVermelha*2;

// Entrada enviada pelo aplicativo
int  state = 0;

void setup()
{

  Serial.begin(9600);
  Serial.println("O bluetooth está ativo. Conecte-se ao HC-05 a partir de qualquer outro dispositivo Bluetooth com 1234 como chave de emparelhamento!");

  // Defina os pinos 2-5 como pinos de saída para acionar a placa de relé
  pinMode(bebida_azul, OUTPUT);
  pinMode(bebida_verde, OUTPUT);
  pinMode(bebida_amarela, OUTPUT);
  pinMode(bebida_vermelha, OUTPUT);


//  Minha placa de relé é uma placa "baixo e alto nível". Portanto, definir a saída para baixa energização do relé. Você não quer que todos liguem ao ligar a coqueteleira...

  digitalWrite(bebida_azul,HIGH);
  digitalWrite(bebida_verde,HIGH);
  digitalWrite(bebida_amarela,HIGH);
  digitalWrite(bebida_vermelha,HIGH);
}

void loop() {

  if(Serial.available()  > 0){ // Checks whether data is comming from the serial port
      state = Serial.read();  // Ler o dado na porta serial
      Serial.println(state);
  }

  Mix();
}

void  Mix() {

  if (state == 'a') {
    Serial.println("Escolhido 250ml de bebida AZUL!");
    digitalWrite(bebida_azul,  LOW);
    delay(_250MlBebidaAzul);
    digitalWrite(bebida_azul, HIGH);
    delay(tempoEntreSuccoes);
  }

  if  (state == 'b') {
    Serial.println("Escolhido 500ml de bebida AZUL!");
    digitalWrite(bebida_azul,  LOW);
    delay(_500MlBebidaAzul);
    digitalWrite(bebida_azul, HIGH);
    delay(tempoEntreSuccoes);
  }

  if (state == 'c') {
    Serial.println("Escolhido 250ml de bebida VERDE!");
    digitalWrite(bebida_verde,  LOW);
    delay(_250MlBebidaVerde);
    digitalWrite(bebida_verde, HIGH);
    delay(tempoEntreSuccoes);
  }

  if  (state == 'd') {
    Serial.println("Escolhido 500ml de bebida VERDE!");
    digitalWrite(bebida_verde,  LOW);
    delay(_500MlBebidaVerde);
    digitalWrite(bebida_verde, HIGH);
    delay(tempoEntreSuccoes);
  }

  if (state == 'e') {
    Serial.println("Escolhido 250ml de bebida AMARELA!");
    digitalWrite(bebida_amarela,  LOW);
    delay(_250MlBebidaAmarela);
    digitalWrite(bebida_amarela, HIGH);
    delay(tempoEntreSuccoes);
  }

  if  (state == 'f') {
    Serial.println("Escolhido 500ml de bebida AMARELA!");
    digitalWrite(bebida_amarela,  LOW);
    delay(_500MlBebidaAmarela);
    digitalWrite(bebida_amarela, HIGH);
    delay(tempoEntreSuccoes);
  }

  if (state == 'g') {
    Serial.println("Escolhido 250ml de bebida VERMELHA!");
    digitalWrite(bebida_vermelha,  LOW);
    delay(_250MlBebidaVermelha);
    digitalWrite(bebida_vermelha, HIGH);
    delay(tempoEntreSuccoes);
  }

  if  (state == 'h') {
    Serial.println("Escolhido 500ml de bebida VERMELHA!");
    digitalWrite(bebida_vermelha,  LOW);
    delay(_500MlBebidaVermelha);
    digitalWrite(bebida_vermelha, HIGH);
    delay(tempoEntreSuccoes);
  }

  if (state == 'i') {
    Serial.println("Escolhido drink de bebida VERMELHA + AZUL!");
    digitalWrite(bebida_vermelha,  LOW);
    delay(_250MlBebidaVermelha);
    digitalWrite(bebida_vermelha, HIGH);
    delay(tempoEntreSuccoes);
    digitalWrite(bebida_azul,  LOW);
    delay(_250MlBebidaAzul);
    digitalWrite(bebida_azul, HIGH);
    delay(tempoEntreSuccoes);
  }

  if  (state == 'j') {
    Serial.println("Escolhido drink de bebida VERMELHA + AZUL + VERDE!");
    // Cada bomba precisa sugar 166ml
    digitalWrite(bebida_vermelha,  LOW);
    delay(_166MlBebidaVermelha);
    digitalWrite(bebida_vermelha, HIGH);
    delay(tempoEntreSuccoes);
    digitalWrite(bebida_azul,  LOW);
    delay(_166MlBebidaAzul);
    digitalWrite(bebida_azul, HIGH);
    delay(tempoEntreSuccoes);
    digitalWrite(bebida_verde,  LOW);
    delay(_166MlBebidaVerde);
    digitalWrite(bebida_verde, HIGH);
    delay(tempoEntreSuccoes);
  }

  if (state == 'k') {
    Serial.println("Escolhido drink de bebida VERMELHA + AMARELO!");
    digitalWrite(bebida_vermelha,  LOW);
    delay(_250MlBebidaVermelha);
    digitalWrite(bebida_vermelha, HIGH);
    delay(tempoEntreSuccoes);
    digitalWrite(bebida_amarela,  LOW);
    delay(_250MlBebidaAmarela);
    digitalWrite(bebida_amarela, HIGH);
    delay(tempoEntreSuccoes);
  }

  if  (state == 'l') {
    Serial.println("Escolhido drink de bebida AMARELO + VERDE!");
    digitalWrite(bebida_amarela,  LOW);
    delay(_250MlBebidaAmarela);
    digitalWrite(bebida_amarela, HIGH);
    delay(tempoEntreSuccoes);
    digitalWrite(bebida_verde,  LOW);
    delay(_250MlBebidaVerde);
    digitalWrite(bebida_verde, HIGH);
    delay(tempoEntreSuccoes);
  }

  // Cleaning
  if (state == 'm') {
    Serial.println("Cleaning");
    digitalWrite(bebida_azul,  LOW);
    delay(tempoLimpeza);
    digitalWrite(bebida_azul, HIGH);
    delay(tempoEntreSuccoes);
    digitalWrite(bebida_verde,  LOW);
    delay(tempoLimpeza);
    digitalWrite(bebida_verde, HIGH);
    delay(tempoEntreSuccoes);
    digitalWrite(bebida_amarela,  LOW);
    delay(tempoLimpeza);
    digitalWrite(bebida_amarela, HIGH);
    delay(tempoEntreSuccoes);
    digitalWrite(bebida_vermelha,  LOW);
    delay(tempoLimpeza);
    digitalWrite(bebida_vermelha, HIGH);
    delay(tempoEntreSuccoes);
  }

  //Serial.println("Execução finalizada!");
  state = 0;
}
```
